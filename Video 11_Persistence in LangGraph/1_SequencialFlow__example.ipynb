#### We will create a sequencial workflow that will generate a joke and then explain a joke on a given topic.

```python
from langgraph.graph import StateGraph, START, END
from langchain_google_genai import ChatGoogleGenerativeAI
from typing import TypedDict
from dotenv import load_dotenv
from langgraph.checkpoint.memory import InMemorySaver
load_dotenv()
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-3.1-flash-lite")
```

```python
class JokeState(TypedDict):
    topic: str
    joke: str
    joke_explanation: str
```

```python
def generate_explanation(state: JokeState) -> JokeState:
    joke = state['joke']
    responce = llm.invoke(f"Give me the explanation and proper context of this joke: {joke}")
    return {"joke_explanation": responce.content[0]['text']}

def generate_joke(state: JokeState) -> JokeState:
    topic = state['topic']
    responce = llm.invoke(f"Create a short funny joke on this topic: {topic}")
    return {"joke": responce.content[0]['text']}
```

```python
graph = StateGraph(JokeState)

## Create Nodes
graph.add_node("generate_joke",generate_joke)
graph.add_node('generate_explanation',generate_explanation)

## Define Edges
graph.add_edge(START, 'generate_joke')
graph.add_edge('generate_joke', 'generate_explanation')
graph.add_edge('generate_explanation', END)

checkpoint = InMemorySaver()
workflow = graph.compile(checkpointer=checkpoint)
```

```python
config_1 = {"configurable": {"thread_id": "thread_1"}}
initial_state_1 = {'topic':"Neet Exam paper Leak"}
final_state_1 = workflow.invoke(initial_state_1, config=config_1)
final_state_1
```

```python
state = workflow.get_state(config=config_1)
state_history = list(workflow.get_state_history(config=config_1))

print(f"This is the current state ourput (data of last execution):\n{state}\n")
print(f"\nThis is the complete history of all executions that took plase:\n{state_history}")
```
