```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
# Define State
class BMIState(TypedDict):
    height: float
    weight: float
    bmi:float
    category: str
```

```python
def calculate_bmi(state: BMIState) -> BMIState:
    height = state['height']
    weight = state['weight']
    
    state['bmi'] = weight/(height**2)
    return state

def healthiness(state: BMIState) -> BMIState:
    bmi = state['bmi']

    if bmi < 18.5:
        state["category"] = "Underweight"
    elif 18.5 <= bmi < 25:
        state["category"] = "Normal"
    elif 25 <= bmi < 30:
        state["category"] = "Overweight"
    else:
        state["category"] = "Obese"
    
    return state
```

```python
graph = StateGraph(BMIState)

# Add node
graph.add_node("calculate_bmi", calculate_bmi)
graph.add_node('healthiness', healthiness)

# Add edges
graph.add_edge(START, 'calculate_bmi')
graph.add_edge('calculate_bmi', 'healthiness')
graph.add_edge('healthiness', END)

# Compile
workflow = graph.compile()
```

```python
final_state = workflow.invoke({"height": 1.6256, "weight": 57})
final_state
```
